```python
# importing libraries
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset
df=pd.read_csv(r'Air India Review.csv')
df
```

```python
# Calculate the average Customer Satisfaction rating
average_rating = df['Rating'].mean()

print(f"Average Customer Satisfaction Rating: {average_rating:.2f}")
```

```python
# Group by the specified columns and calculate the mean rating for each group
grouped_ratings = df.groupby([
    'Detail.Type Of Traveller',
    'Detail.Seat Type',
    'Detail.Route',
    'Detail.Date Flown',
    'Detail.Seat Comfort',
    'Detail.Cabin Staff Service',
    'Detail.Food & Beverages',
    'Detail.Inflight Entertainment',
    'Detail.Ground Service',
    'Detail.Wifi & Connectivity',
    'Detail.Value For Money'
])['Rating'].mean().reset_index()

# Display the grouped ratings
grouped_ratings
```

```python
# Most common route
common_route = df['Detail.Route'].mode()[0]

# Count of the most common route
common_route_count = df['Detail.Route'].value_counts().max()
print(f"The most common route is '{common_route}' and appears '{common_route_count}' times in the dataset.")
```

```python
# Counting occurrences of the word "refund"
refund_mentions = df['Review_Body'].str.contains('refund', case=False, na=False).sum()
print(f"Number of reviews mentioning 'refund': {refund_mentions}")
print('-----------------------------------------------------------------')
# Counting occurrences of the word "Trip Verified"
TV_mentions = df['Review_Body'].str.contains('Trip Verified', case=False, na=False).sum()
print(f"Number of reviews mentioning 'Trip Verified': {TV_mentions}")
```

```python
# Define sentiment categories
def categorize_sentiment(rating):
    if rating >= 5:
        return 'Positive'
    elif rating<5 and rating >=3:
        return 'Neutral'
    else:
        return 'Negative'

df['Sentiment'] = df['Rating'].apply(categorize_sentiment)

# Calculate sentiment distribution
sentiment_distribution = df['Sentiment'].value_counts(normalize=True) * 100
print(sentiment_distribution)

# Plot sentiment distribution
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x=sentiment_distribution.index, y=sentiment_distribution.values)
plt.title('Sentiment Distribution')
plt.ylabel('Percentage')
plt.show()
```

```python
# Average rating for different types of aircraft
avg_rating_aircraft = df.groupby('Detail.Aircraft')['Rating'].mean()
print(avg_rating_aircraft)
```

```python
# Top 5 Most Common Routes in the Dataset
top_routes = df['Detail.Route'].value_counts().nlargest(5)
print(top_routes)
```

```python
#top 10 routes with the highest average ratings
avg_rating_by_route = df.groupby('Detail.Route')['Rating'].mean()

top_10_routes = avg_rating_by_route.sort_values(ascending=False).head(10)
print(top_10_routes)

```

```python
#Correlation Between Different Service Ratings and Overall Rating
correlation_matrix = df[['Rating', 'Detail.Cabin Staff Service', 'Detail.Food & Beverages',
                         'Detail.Inflight Entertainment', 'Detail.Ground Service',
                         'Detail.Wifi & Connectivity', 'Detail.Value For Money']].corr()
correlation_matrix
```

```python
#Top 10 Routes with Highest Average Value for Money Rating
top_routes_value_for_money = df.groupby('Detail.Route')['Detail.Value For Money'].mean().nlargest(10)
print(top_routes_value_for_money)
```

```python
# Average rating by seat comfort
avg_rating_seat_comfort = df.groupby('Detail.Seat Comfort')['Rating'].mean()
print(avg_rating_seat_comfort)
```

```python
from wordcloud import WordCloud

# Define a list of negative words
negative_words = ['pathetic', 'worst', 'horrible', 'terrible', 'poor', 'disappointing', 'bad', 'awful', 'sucks', 'unacceptable','dirty','poor']

# Function to extract negative words from review body
def extract_negative_words(review_body):
    review_body = review_body.lower()  # Convert to lowercase for consistency
    found_words = [word for word in negative_words if word in review_body]
    return ' '.join(found_words)  # Join found words into a single string

# Apply the function to extract negative words from each review
df['Extracted_Negative_Words'] = df['Review_Body'].apply(extract_negative_words)

# Combine all extracted negative words into a single string
all_negative_words = ' '.join(df['Extracted_Negative_Words'])

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_negative_words)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide axis
plt.show()
```

```python
from collections import Counter

# Function to extract negative words from review body
def extract_negative_words(review_body):
    review_body = review_body.lower()  # Convert to lowercase for consistency
    found_words = [word for word in negative_words if word in review_body]
    return found_words

# Apply the function to extract negative words from each review
df['Extracted_Negative_Words'] = df['Review_Body'].apply(extract_negative_words)

# Flatten the list of extracted negative words into a single list
all_negative_words = [word for sublist in df['Extracted_Negative_Words'] for word in sublist]

# Count the occurrences of each negative word
negative_word_counts = Counter(all_negative_words)

# Display the counts
for word, count in negative_word_counts.items():
    print(f"{word}: {count}")
```

```python
# checks the flight recommendation
df['Detail.Recommended'] = df['Detail.Recommended'].str.lower()  # Convert to lowercase for consistency

# Count the number of occurrences of 'yes' and 'no'
recommendation_counts = df['Detail.Recommended'].value_counts()

# Display the results
print(f"Number of recommended flights: {recommendation_counts.get('yes', 0)}")
print(f"Number of not recommended flights: {recommendation_counts.get('no', 0)}")
```
